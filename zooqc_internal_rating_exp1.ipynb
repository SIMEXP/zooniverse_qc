### EXP1 : internal validation 

- In this EXP I choose 30 maybe, 8 OK and 12 fail subjects for ADHD200 total of 50 subjects
- Pierre and I Rated them 1 time each

```python
# Load library
import pandas as pd
import os
```

```python
# paths
path_root = "/home/yassinebha/Drive"
path_qc = os.path.join(path_root,"QC_zooniverse/zooqc_internal_validation")
```

```python
# read CSV's
qc_pierre = pd.read_csv(os.path.join(path_qc,"intarnal_validation_exp1_Pierre_rating.csv"))
qc_yassine = pd.read_csv(os.path.join(path_qc,"intarnal_validation_exp1_Yassine_rating.csv"))
subject_match_id = pd.read_csv(os.path.join(path_qc,"brain-match-subjects_exp1_internal_rating.csv"))
qc_pierre.head()
```

```python
qc_yassine.head()
```

```python
subject_match_id.head()
```

```python
# retrive subject ID from the list of string in each cell
subjects = subject_match_id.metadata.str[-10:-2]
subject_match_id.loc[:,'ID']=subjects
subject_match_id.head()
```

```python
# Add subject id Correspondance for Pierre
qc_pierre_id = pd.merge(qc_pierre,subject_match_id[["subject_id","ID"]], how = 'left', on = 'subject_id')
qc_pierre_id.head()
```

```python
# Add subject id Correspondance for Yassine
qc_yassine_id = pd.merge(qc_yassine,subject_match_id[["subject_id","ID"]], how = 'left', on = 'subject_id')
qc_yassine_id.head()
```

```python
# merge yassine pierre csv's
merged_qc = pd.merge(qc_pierre_id,qc_yassine_id,on='ID',how='left')
merged_qc
```

```python
# Replace OK:1 , Maybe:2 and Fail:3
merged_qc.replace({'OK':1,'Maybe':2, 'Fail':3}, inplace=True)
merged_qc.head()
```

```python
from sklearn import metrics
kappa = metrics.cohen_kappa_score(merged_qc[['most_likely_x']], merged_qc[['most_likely_y']])
print("The kappa betwen Yassine and Pierre is %f " %(kappa))
```
