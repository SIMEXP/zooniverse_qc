### EXP1a : internal validation 

- In this EXP I choose 16 OK from COBRE and 17 fail, 17 maybe  from ADHD200 total of 50 subjects
- Pierre and I Rated them **2 times** each
- I'll compute kappa intra and inter rater

```python
# Load library
import pandas as pd
import os
import numpy as np
```

```python
# paths
path_root = "/media/yassinebha/database29/Drive"
path_qc = os.path.join(path_root,"QC_zooniverse/zooqc_internal_validation/EXP1a")
```

### Calculate Kappa intra rater
---------------------------------
** PIERRE **

```python
# read CSV's
qc_pierre_1 = pd.read_csv(os.path.join(path_qc,"pierre/pierre_1/T0Rate_the_match_summary.csv"))
qc_pierre_2 = pd.read_csv(os.path.join(path_qc,"pierre/pierre_2/T0Rate_the_match_summary.csv"))
```

```python
qc_pierre_1[["subject_id","most_likely"]].head()
```

```python
qc_pierre_2[["subject_id","most_likely"]].head()
```

```python
# match SubjID with zooniverse ID for Pierre_1 and 2
zoo_id = pd.read_csv(os.path.join(path_qc,"brain-match-subjects_exp1a_raw.csv"))
zoo_id[["subject_id","id"]].head()
```

```python
# Add subject id Correspondance for Pierre_1
qc_pierre_1_id = pd.merge(qc_pierre_1,zoo_id[["subject_id","id"]], how = 'left', on = 'subject_id')
qc_pierre_1_id[["id","most_likely","subject_id"]].head()
```

```python
# Add subject id Correspondance for Pierre_2
qc_pierre_2_id = pd.merge(qc_pierre_2,zoo_id[["subject_id","id"]], how = 'left', on = 'subject_id')
qc_pierre_2_id[["id","most_likely","subject_id"]].head()
```

```python
# Merge Pierre_1 and 2
qc_pierre_1_2 = pd.merge(qc_pierre_1_id[["id","most_likely"]],
                         qc_pierre_2_id[["id","most_likely"]], 
                         how = 'left', on = "id")
qc_pierre_1_2.loc[:,"rating_match_pierre"] = [ row["most_likely_x"] == row["most_likely_y"] for ind,row in qc_pierre_1_2.iterrows()]
qc_pierre_1_2
```

```python
# save pierre's rating to variable for later saving
df_save_pierre = qc_pierre_1_2[['id','most_likely_x','most_likely_y']].rename(columns={'id':'ID','most_likely_x':'rating_pierre_1','most_likely_y':'rating_pierre_2'})
df_save_pierre.head()
```

```python
# number of mismatch intra rater
50 - np.sum(qc_pierre_1_2["rating_match_pierre"])
```

```python
# Replace OK:1 , Maybe:2 and Fail:3
qc_pierre_1_2.replace({'OK':1,'Maybe':2, 'Fail':3}, inplace=True)
qc_pierre_1_2.head()
```

```python
# Calculate kappa for pierre 1 and 2
from sklearn import metrics
kappa = metrics.cohen_kappa_score(qc_pierre_1_2[['most_likely_x']],qc_pierre_1_2[['most_likely_y']])
print("The kappa for Pierre is %f " %(kappa))
```
